```python
#dataset2
import requests
import pandas as pd


url = 'https://www.ffiec.gov/census/report.aspx?year=2021&county=037&state=06&report=demographic'
html = requests.get(url).content
df_list = pd.read_html(html)
df = df_list[-1]

df.to_csv('my data.csv')
df = pd.read_csv("my data.csv", usecols = [4,6,7], nrows = 50)  

df
```

```python
df['2021 Est. Tract Median Family Income'] = pd.to_numeric(df['2021 Est. Tract Median Family Income'].str.replace('$','').str.replace(',',''))
```

```python
df['2021 Est. Tract Median Family Income']
```

```python
df['2015 Tract Median Family Income'] = pd.to_numeric(df['2015 Tract Median Family Income'].str.replace('$','').str.replace(',',''))
```

```python
df.to_csv('my data.csv')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 10))
x = range(0, 50)
y_1 = df['2021 Est. Tract Median Family Income']
y_2 = df['2015 Tract Median Family Income']

plt.xlabel('Families')
plt.ylabel('Family monthly incomes(dollars)')

plt.title("Comparison of sample monthly income of LA citizens")

plt.plot(x, y_1, color='blue', label='2021', linewidth=3.0, marker='.')
plt.plot(x, y_2, color='orange', label='2015', linewidth=3.0, marker='x')

plt.legend(['2021', '2015'])
 

plt.show()
```

```python

import requests
import json
import pandas as pd

America_url = 'https://view.inews.qq.com/g2/getOnsInfo?name=disease_foreign'
headers = {
'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36'
,'referer': 'https://news.qq.com/'
}

response = requests.get(url=America_url,headers = headers).json()


data = json.loads(response['data'])


with open('./USCOVID.json','w') as f:
   
    f.write(json.dumps(data,indent=2,ensure_ascii=False))   #indent为格式空两格的操作


with open('./USCOVID.json','r') as f:
    data = f.read()


data = json.loads(data)


AmericaDict = data['foreignList'][0]


ProvinceList = AmericaDict['children']


AmericaCityList = []
AmericaTotal = {'Date':AmericaDict['date'],'States':'North America','Province':'America','Confirmed':AmericaDict['confirm'],'Heal':AmericaDict['heal'],'Death':AmericaDict['dead']}
AmericaCityList.append(AmericaTotal)


for i in range(len(ProvinceList)):
    date = ProvinceList[i]['date']
    province_English = ProvinceList[i]['nameMap']
    province = ProvinceList[i]['name']
    dead = ProvinceList[i]['dead']
    heal = ProvinceList[i]['heal']
    confirm = ProvinceList[i]['confirm']
    city_list = {'Date':date,'States':province_English,'Province':province,'Confirmed':confirm,'Heal':heal,'Death':dead}
    AmericaCityList.append(city_list)


AmericaTotalData = pd.DataFrame(AmericaCityList)
```
